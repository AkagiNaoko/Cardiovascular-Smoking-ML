```python
import pandas as pd
from pathlib import Path
```

```python
# Read in the CSV file as a Pandas DataFrame
tobacco_df = pd.read_csv(
    Path("Resources\Data\Tobacco_Behavioral_Risk_Factor_Data__Tobacco_Use__2011_to_present_.csv")
)

# Review the DataFrame
tobacco_df.head()
```

```python
print(tobacco_df.info())
```

```python
# Check for missing values
print(tobacco_df.isnull().sum())
```

```python
# Remove columns that are not necessary for the project objective
columns_to_remove = ['LocationDesc', 'TopicType', 'DataSource', 'Response', 'Data_Value_Unit', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'Data_Value_Std_Err', 'GeoLocation', 'TopicTypeId', 'TopicId', 'MeasureId', 'StratificationID1', 'StratificationID2', 'StratificationID3', 'StratificationID4', 'SubMeasureID', 'DisplayOrder', 'States']
tobacco_data_cleaned = tobacco_df.drop(columns=columns_to_remove, axis=1)

# Display the DataFrame after dropping columns
tobacco_data_cleaned
```

```python
# Check for missing values again
print(tobacco_data_cleaned.isnull().sum())
```

```python
# Fill missing values in numerical columns with median
tobacco_data_cleaned['Data_Value'].fillna(tobacco_data_cleaned['Data_Value'].median(), inplace=True)
tobacco_data_cleaned['Low_Confidence_Limit'].fillna(tobacco_data_cleaned['Low_Confidence_Limit'].median(), inplace=True)
tobacco_data_cleaned['High_Confidence_Limit'].fillna(tobacco_data_cleaned['High_Confidence_Limit'].median(), inplace=True)
tobacco_data_cleaned['Sample_Size'].fillna(tobacco_data_cleaned['Sample_Size'].median(), inplace=True)

# Check for missing values again
print(tobacco_data_cleaned.isnull().sum())
```

```python
# Standardize categorical values by converting to lowercase and removing leading/trailing spaces
tobacco_data_cleaned['TopicDesc'] = tobacco_data_cleaned['TopicDesc'].str.strip().str.lower()
tobacco_data_cleaned['MeasureDesc'] = tobacco_data_cleaned['MeasureDesc'].str.strip().str.lower()
tobacco_data_cleaned['Gender'] = tobacco_data_cleaned['Gender'].str.strip().str.lower()
tobacco_data_cleaned['Race'] = tobacco_data_cleaned['Race'].str.strip().str.lower()
tobacco_data_cleaned['Age'] = tobacco_data_cleaned['Age'].str.strip().str.lower()
tobacco_data_cleaned['Education'] = tobacco_data_cleaned['Education'].str.strip().str.lower()

# Merge similar categories in 'Race' column
tobacco_data_cleaned['Race'] = tobacco_data_cleaned['Race'].replace({
    'all races': 'all races/ethnicities',
    'black, non-hispanic': 'black',
    'white, non-hispanic': 'white',
    'asian, non-hispanic': 'asian',
    'hispanic': 'hispanic/latino',
    'other race, non-hispanic': 'other'
})

# Display unique values in 'Race' column after merging
print(tobacco_data_cleaned['Race'].unique())
```

```python
# Merge similar categories in 'Education' column
tobacco_data_cleaned['Education'] = tobacco_data_cleaned['Education'].replace({
    'all grades': 'all education levels',
    '12th grade (including ged)': 'high school graduate or ged',
    '< 12th grade': 'less than high school'
})

# Display unique values in 'Education' column after merging
print(tobacco_data_cleaned['Education'].unique())
```
