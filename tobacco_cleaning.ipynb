```python
status_df= tobacco_df.loc[(tobacco_df["MeasureDesc"].isin(["Smoking Status"]))]
status_df
```

```python
status_df['YearStart'] = status_df['YEAR'].astype(str)
status_df['Year-State'] = status_df['YEAR'] + '-' + status_df['LocationAbbr']
status_df.head()
```

```python
year_state_list = []
current_overall = []
current_female = []
current_male =[]
Never_overall =[]
Never_female = []
Never_male =[]
Former_overall =[]
Former_female =[]
Former_male = []
for index,row in status_df.iterrows():
    year_state = row['Year-State']
    if year_state not in year_state_list:
        year_state_list.append(year_state)
    if row['Response'] == "Current":
        if row['Gender'] == 'Overall':        
            value = row['Data_Value'] 
            current_overall.append(value)
        elif row['Gender'] == 'Male':
            value = row['Data_Value'] 
            current_male.append(value)
        else:
            value = row['Data_Value'] 
            current_female.append(value)
    elif row['Response'] == "Never":
        if row['Gender'] == 'Overall':        
            value = row['Data_Value'] 
            Never_overall.append(value)
        elif row['Gender'] == 'Male':
            value = row['Data_Value'] 
            Never_male.append(value)
        else:
            value = row['Data_Value'] 
            Never_female.append(value)
    else:
        if row['Gender'] == 'Overall':        
            value = row['Data_Value'] 
            Former_overall.append(value)
        elif row['Gender'] == 'Male':
            value = row['Data_Value'] 
            Former_male.append(value)
        else:
            value = row['Data_Value'] 
            Former_female.append(value)
```

```python
Cigaretee_df = pd.DataFrame({"Year-State": year_state_list,
                        "current_overall": current_overall,
                       "current_female": current_female, 
                       "current_male":current_male,
                       "Never_overall": Never_overall,
                       "Never_female": Never_female,
                       "Never_male":Never_male,
                       "Former_overall":Former_overall,
                       "Former_female": Former_female,
                       "Former_male": Former_male}) 
Cigaretee_df
```

```python
Cigaretee_df.to_csv('cigaretee.csv', index=False)
```
