```python
import pandas as pd
from pathlib import Path
```

```python
# Read in the CSV file as a Pandas DataFrame
cardio_df = pd.read_csv(
    Path("Cardio_Behavioral_Risk_Factor_Surveillance_System__BRFSS__-__National_Cardiovascular_Disease_Surveillance_Data.csv")
)

# Review the DataFrame
cardio_df.head()
```

```python
# remove crude prevelance
cardio_df = cardio_df.loc[(cardio_df["Data_Value_Type"] == "Age-Standardized"),:]
cardio_df
```

```python
cardio_df["Break_Out"].value_counts()
```

```python
#remove US mean rows
cardio_df = cardio_df[cardio_df["LocationAbbr"] != 'USM']
cardio_df
```

```python
cardio_df = cardio_df[cardio_df["YearStart"] != 2020]
cardio_df
```

```python
cardio_df = cardio_df.drop(columns = ["RowId","DataSource", "Class","Topic","LocationDesc", "PriorityArea1", "PriorityArea2", "PriorityArea3", "PriorityArea4", "Data_Value_Alt", "Data_Value_Unit", "Data_Value_TypeID", "Data_Value_Footnote_Symbol", "ClassId","TopicId","QuestionId","BreakOutCategoryId", "BreakOutId","LocationId", "Geolocation", "Data_Value_Footnote", "Low_Confidence_Limit", "High_Confidence_Limit"])
cardio_df
```

```python
cardio_df = cardio_df[cardio_df['Data_Value'].notna()]
cardio_df
```

```python
cardio_df = cardio_df[cardio_df['Break_Out_Category'].isin(['Overall','Gender'])]
cardio_df
```

```python
cardio_df["Question"].value_counts()
```

```python
cardio_df = cardio_df[cardio_df['Question'].isin(['Prevalence of major cardiovascular disease among US adults (18+); BRFSS',
                                                  'Prevalence of acute myocardial infarction (heart attack) among US adults (18+); BRFSS',
                                                 'Prevalence of coronary heart disease among US adults (18+); BRFSS',
                                                 'Prevalence of stroke among US adults (18+); BRFSS',
                                                 'Prevalence of diabetes among US adults (18+); BRFSS',
                                                 'Prevalence of obesity among US adults (20+); BRFSS',
                                                 'Prevalence of physical inactivity among US adults (18+); BRFSS',
                                                 'Prevalence of physical inactivity among US adults (18+); BRFSS',
                                                 'Prevalence of physical inactivity among US adults (18+); BRFSS'])]
cardio_df
```

```python
cardio_df["Question"].value_counts()
```

```python
cardio_df['YearStart'] = cardio_df['YearStart'].astype(str)
cardio_df['Year-State'] = cardio_df['YearStart'] + '-' + cardio_df['LocationAbbr']
cardio_df.head()
```

```python
year_state_list = []
for index,row in cardio_df.iterrows():
    year_state = row['Year-State']
    if year_state not in year_state_list:
        year_state_list.append(year_state)
```

```python
major_cvd_overall = []
major_cvd_male = []
major_cvd_female = []

for index,row in cardio_df.iterrows():
    if row['Question'] == "Prevalence of major cardiovascular disease among US adults (18+); BRFSS":
        if row['Break_Out'] == 'Overall':        
            value = row['Data_Value'] 
            major_cvd_overall.append(value)
        elif row['Break_Out'] == 'Male':
            value = row['Data_Value'] 
            major_cvd_male.append(value)
        else:
            value = row['Data_Value'] 
            major_cvd_female.append(value)
```
